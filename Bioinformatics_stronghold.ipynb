Mendel's First Law

Given: Three positive integers k, m, and n, representing a population containing k+m+n organisms: k individuals are homozygous dominant for a factor, m are heterozygous, and n are homozygous recessive.

Return: The probability that two randomly selected mating organisms will produce an individual possessing a dominant allele (and thus displaying the dominant phenotype). Assume that any two organisms can mate.

```python
def mendel_first_law(k,m,n):
    kmn = k+m+n
    k_k = 1   * k/kmn*(k-1)/(kmn-1)
    m_m = 3/4 *(m/kmn * (m-1)/(kmn-1))
    n_n = 0
    
    k_m = 1   *k/kmn*m/(kmn-1) + 1   *m/kmn*k/(kmn-1)
    k_n = 1   *k/kmn*n/(kmn-1) + 1   *n/kmn*k/(kmn-1)
    m_n = 1/2 *m/kmn*n/(kmn-1) + 1/2 *n/kmn*m/(kmn-1)
    
    sum = k_k + m_m + n_n + k_m + k_n + m_n
    return sum

k, m, n = (int(i) for i in (input().split()))
print(mendel_first_law(k,m,n))    
```

Mortal Fibonacci Rabbits

Given: Positive integers n≤100 and m≤20.

Return: The total number of pairs of rabbits that will remain after the n-th month if all rabbits live for m months.

```python
n, m = (int(i) for i in input().split())

def one_month_later(rabits):
    s_new = sum(rabits[1:])
    for i in reversed(range(1,m)):
        rabits[i] = rabits[i-1]
    rabits[0] = s_new

rabits = [0]*m
rabits[0] = 1

print(rabits)

for i in range(1,n):
    one_month_later(rabits)
    print(rabits)
    
print(sum(rabits))
```

Independent Alleles

Given: Two positive integers k (k≤7) and N (N≤2k). In this problem, we begin with Tom, who in the 0th generation has genotype Aa Bb. Tom has two children in the 1st generation, each of whom has two children, and so on. Each organism always mates with an organism having genotype Aa Bb.

Return: The probability that at least N Aa Bb organisms will belong to the k-th generation of Tom's family tree (don't count the Aa Bb mates at each level). Assume that Mendel's second law holds for the factors.

```python
k, N = (int(i) for i in input().split())

import math

def p_n(n, k):
    C_2k_n = (math.factorial(2**k)/(math.factorial(n)*math.factorial((2**k)-n)))
    P_n = (C_2k_n*(0.25**n)*(0.75**(2**k - n)))
    return P_n

probability = 0
for i in range(N, 2**k+1):
    probability += p_n(i, k)
    
print(probability)
```

Finding a Protein Motif

Given: At most 15 UniProt Protein Database access IDs.

Return: For each protein possessing the N-glycosylation motif N{P}[ST]{P}, output its given access ID followed by a list of locations in the protein string where the motif can be found.

```python
import requests
import re

def get_a_string(new_link):
    page = requests.get(new_link)
    file = page.text
    string = file.strip().split('\n')
    new_string = ''
    for el in string[1:]:
        new_string += el
    return new_string

def find_a_pattern(new_string, pattern):
    pos = []
    k = 0
    while k < len(new_string):
        m = re.search(pattern, new_string[k:])
        if m:
            place = m.span()[0] + 1
            k += place
            pos.append(k)
        else:
            break
    return pos
        
name_loc = {}
pattern = r'N[^P][ST][^P]'
        
with open('inf.txt') as inf:
    for line in inf:
        new_link = 'http://www.uniprot.org/uniprot/' + line.strip() + '.fasta'
        new_string = get_a_string(new_link)
        positions = find_a_pattern(new_string, pattern)
        name_loc[line.strip()] = positions      

with open('ouf.txt', 'w') as ouf:
    for key in name_loc:
        if len(name_loc[key]) != 0:
            ouf.write(key)
            ouf.write('\n')
            ouf.write(' '.join(str(i) for i in name_loc[key]))
            ouf.write('\n')
```

Open Reading Frames

Given: A DNA string s of length at most 1 kbp in FASTA format.

Return: Every distinct candidate protein string that can be translated from ORFs of s. Strings can be returned in any order.

```python
DNA_codon_table = {
'TTT': 'F',      'CTT': 'L',      'ATT': 'I',      'GTT': 'V',
'TTC': 'F',      'CTC': 'L',      'ATC': 'I',      'GTC': 'V',
'TTA': 'L',      'CTA': 'L',      'ATA': 'I',      'GTA': 'V',
'TTG': 'L',      'CTG': 'L',      'ATG': 'M',      'GTG': 'V',
'TCT': 'S',      'CCT': 'P',      'ACT': 'T',      'GCT': 'A',
'TCC': 'S',      'CCC': 'P',      'ACC': 'T',      'GCC': 'A',
'TCA': 'S',      'CCA': 'P',      'ACA': 'T',      'GCA': 'A',
'TCG': 'S',      'CCG': 'P',      'ACG': 'T',      'GCG': 'A',
'TAT': 'Y',      'CAT': 'H',      'AAT': 'N',      'GAT': 'D',
'TAC': 'Y',      'CAC': 'H',      'AAC': 'N',      'GAC': 'D',
'TAA': 'Stop',   'CAA': 'Q',      'AAA': 'K',      'GAA': 'E',
'TAG': 'Stop',   'CAG': 'Q',      'AAG': 'K',      'GAG': 'E',
'TGT': 'C',      'CGT': 'R',      'AGT': 'S',      'GGT': 'G',
'TGC': 'C',      'CGC': 'R',      'AGC': 'S',      'GGC': 'G',
'TGA': 'Stop',   'CGA': 'R',      'AGA': 'R',      'GGA': 'G',
'TGG': 'W',      'CGG': 'R',      'AGG': 'R',      'GGG': 'G'
}

with open('inf.txt') as inf:
    s_DNA = ''
    for line in inf:
        if line[0] != '>':
            s_DNA += line.strip()

start_DNA = 'ATG'
stop_DNA = ['TAA', 'TAG', 'TGA']


DNA_seq = []
def make_DNA_sequences(s_DNA):

    try_start = []
    try_stop = []

    i = 0
    while i < len(s_DNA)-3:
        if s_DNA[i:i+3] == start_DNA:
            try_start.append(i)
        for el in stop_DNA:
            if s_DNA[i:i+3] == el:
                try_stop.append(i)
        i += 1

    
    for start_loc in try_start:
        for stop_loc in try_stop:
            if stop_loc > start_loc and (stop_loc-start_loc) % 3 == 0:
                DNA_seq.append(s_DNA[start_loc:stop_loc])
                break

            
def translate_DNA_to_protein(DNA_seq):
    protein = ''
    for i in range(0, len(DNA_seq), 3):
        cod = DNA_seq[i:i+3]
        protein += DNA_codon_table[cod]
    return protein
        
    
complements = {
    'A' : 'T',
    'T' : 'A',
    'C' : 'G',
    'G' : 'C'
}
def reverse_complement(dna):
    dna_new = ''
    for i in reversed(range(len(dna))):
        dna_new += complements[dna[i]]
    return dna_new

                

make_DNA_sequences(s_DNA)
dna_reversed = reverse_complement(s_DNA)
make_DNA_sequences(dna_reversed)

with open('ouf.txt', 'w') as ouf:
    for seq in set(DNA_seq):
        protein = translate_DNA_to_protein(seq)
        ouf.write(protein)
        ouf.write('\n')
```

Genome Assembly as Shortest Superstring

Given: At most 50 DNA strings of approximately equal length, not exceeding 1 kbp, in FASTA format (which represent reads deriving from the same strand of a single linear chromosome).

The dataset is guaranteed to satisfy the following condition: there exists a unique way to reconstruct the entire chromosome from these reads by gluing together pairs of reads that overlap by more than half their length.

Return: A shortest superstring containing all the given strings (thus corresponding to a reconstructed chromosome).

```python
# make function to find suffix length
def find_suffix_length(a,b):
    suf_len = 0
    min_len = min(len(a), len(b))
    for i in range(min_len//2, min_len+1):
        if a[-i:] == b[:i]:
            suf_len = i
            break
    return suf_len


# make a reads list from file
name_DNA = {}
with open('inf.txt') as inf:
    for line in inf:
        line = line.strip()
        if line[0] == '>':
            name = line[1:]
            name_DNA[name] = ''
        else:
            name_DNA[name] += line

reads = list(name_DNA.values())
l = min(map(len, reads))//2
n = len(reads)


# make matrix of suffix length for reads
len_matrix = [[0 for x in range(n)] for y in range(n)]
for i, el_i in enumerate(reads):
    for j, el_j in enumerate(reads):
        if i == j:
            len_matrix[i][j] = -1
            continue
        len_matrix[i][j] = find_suffix_length(el_i,el_j)


# find possible pairs
candidate_cells = [[(i,j) for j in range(n) if len_matrix[i][j] >= l] for i in range(n)]

# find last string
for i in range(len(candidate_cells)):
    if len(candidate_cells[i]) == 0:
        last_ind = i
              
# make list of row numbers in right order
sequence = [last_ind]
while len(sequence) < len(reads):
    for i in range(len(candidate_cells)):
        if i != last_ind and candidate_cells[i][0][1] == sequence[-1]:
            sequence.append(candidate_cells[i][0][0])
            break        
sequence = sequence[::-1]
row_numbers = sequence


# combine strings to superstring
shortest_superstring = ''
for i in range(0, len(row_numbers)-1):
    number = row_numbers[i]
    next_number = row_numbers[i+1]
    end_of_string = len_matrix[number][next_number]
    shortest_superstring += reads[number][:-end_of_string]
shortest_superstring += reads[row_numbers[-1]]


# write answer to file
with open('ouf.txt', 'w') as ouf:
    ouf.write(shortest_superstring)
```

Perfect Matchings and RNA Secondary Structures

Given: An RNA string s of length at most 80 bp having the same number of occurrences of 'A' as 'U' and the same number of occurrences of 'C' as 'G'.

Return: The total possible number of perfect matchings of basepair edges in the bonding graph of s.

```python
# read RNA from file
RNA = ''
with open('inf.txt') as inf:
    for line in inf:
        line = line.strip()
        if line[0] == '>':
            continue
        else:
            RNA += line
            
#print(RNA)

# recursion relation
def total_number_perfect_matchings(n):
    if n == 0:
        return 1
    else:
        return (n)*total_number_perfect_matchings(n-1)


# count perfect edges
n1 = RNA.count('A')
basepair_edges_AU = total_number_perfect_matchings(n1)
n2 = RNA.count('C')
basepair_edges_CG = total_number_perfect_matchings(n2)
perfect_basepair_edges = basepair_edges_AU * basepair_edges_CG

print(perfect_basepair_edges)
```

Introduction to Random Strings

Given: A DNA string s of length at most 100 bp and an array A containing at most 20 numbers between 0 and 1.

Return: An array B having the same length as A in which B[k] represents the common logarithm of the probability that a random string constructed with the GC-content found in A[k] will match s exactly.

```python
import math

def find_log_prob(dna, gc_content):
    symb_prob = {}
    symb_prob['C'] = gc_content/2
    symb_prob['G'] = gc_content/2
    symb_prob['A'] = (1 - gc_content)/2
    symb_prob['T'] = (1 - gc_content)/2
    log_prob = 0
    for symb in dna:
        log_prob += math.log10(symb_prob[symb])
    return log_prob


s = input()
A = list(map(float, input().strip().split()))

B = [0 for i in range(len(A))]

for i in range(len(A)):
    B[i] = (find_log_prob(s, A[i]))
    
print(*[f'{x:0.3f}' for x in B])
```

Enumerating Oriented Gene Orderings

Given: A positive integer n≤6.

Return: The total number of signed permutations of length n, followed by a list of all such permutations (you may list the signed permutations in any order).

```python
n = int(input())
pairs = []

import itertools

numbers = list(range(1, n+1))
permutations = list(itertools.permutations(numbers))
signs_comb = list(itertools.product([1,-1],repeat = n))

 
for item_1 in permutations:
    for item_2 in signs_comb:
        s = ''
        for i in range(n):
            s += str(item_1[i] * item_2[i]) + ' '
        pairs.append(s)
        
with open('ouf.txt', 'w') as ouf:
    ouf.write(str(len(pairs))+'\n')
    ouf.write('\n'.join(pairs))
```

Transitions and Transversions

Given: Two DNA strings s1 and s2 of equal length (at most 1 kbp).

Return: The transition/transversion ratio R(s1,s2).

```python
name_DNA = {}
with open('inf.txt') as inf:
    for line in inf:
        line = line.strip()
        if line[0] == '>':
            name = line[1:]
            name_DNA[name] = ''
        else:
            name_DNA[name] += line

dna_list = list(name_DNA.values())

transition = {'A':'G', 'G':'A', 'C':'T', 'T':'C'}


def transition_transversion_ratio(s1, s2):
    n = len(s1)
    transi = 0
    transv = 0
    for i in range(n):
        if s1[i] == s2[i]:
            continue
        elif transition[s1[i]] ==  s2[i]:
            transi += 1
        else:
            transv += 1
    R = transi/transv
    return R

print(transition_transversion_ratio(dna_list[0], dna_list[1]))
```

Completing a Tree

Given: A positive integer n (n≤1000) and an adjacency list corresponding to a graph on n nodes that contains no cycles.

Return: The minimum number of edges that can be added to the graph to produce a tree.

```python
with open('inf.txt') as inf:
    n = int(inf.readline())
    nodes = {}
    for i in range(1, n+1):
        nodes[i] = []
    while True:
        line = inf.readline().strip().split()
        if not line:
            break
        x, y = (map(int, line))
        nodes[x].append(y)
        nodes[y].append(x)
```

```python
def find_cluster(nodes, current_node, cluster=set()):
#     print(cluster)
    cluster.add(current_node)
    children = [c for c in nodes[current_node] if c not in cluster]
    if len(children) == 0:
        return cluster
    else:
        for c in children:
            cluster.update(find_cluster(nodes, c, cluster))
        return cluster
```

```python
used_nodes = set()
n_clusters = 0
for node in nodes.keys():
    if node not in used_nodes:
        cluster = find_cluster(nodes, node, cluster=set())
        used_nodes.update(cluster)
        n_clusters += 1
        
print(n_clusters - 1)
```
